# New Data

```python
data = data_raw[data_raw.notnull()]
```

```python
data.isnull().sum().sum()
```

```python
data[['Name','Age','Potential']][data.notnull()]
```

```python
d1 = data[['Name','Age','Potential','Overall','Vision']][data.Age==19]
d2 = data[['Name','Age','Potential','Overall','Vision']][data.Age==25]
```

```python
fig=plt.figure() #Plots in matplotlib reside within a figure object, use plt.figure to create new figure
#Create one or more subplots using add_subplot, because you can't create blank figure
ax = fig.add_subplot(1,1,1)
#Variable
ax.hist(d1['Vision'],bins = 5) # Here you can play with number of bins
plt.title('Age 19')
plt.xlabel('V')
plt.ylabel('Number of Players')
plt.show()
```

```python
sns.violinplot(data['Age'], data['Vision']) #Variable Plot
sns.despine()
```

```python
fig=plt.figure() #Plots in matplotlib reside within a figure object, use plt.figure to create new figure
#Create one or more subplots using add_subplot, because you can't create blank figure
ax = fig.add_subplot(1,1,1)
#Variable
ax.hist(d2['Vision'],bins = 5) # Here you can play with number of bins
plt.title('Age 25')
plt.xlabel('V')
plt.ylabel('Number of Players')
plt.show()
```
